# 13th session: Taylor Approximation

$$
f(x) \approx f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f'''(a)}{3!}(x - a)^3 + \cdots + \frac{f^{(n)}(a)}{n!}(x - a)^n
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
range_of_x = np.linspace(-10, 10, 1000)
y = np.e ** range_of_x

plt.plot(range_of_x, y, label='e^x')

plt.xlim(-1, 1)
plt.ylim(0, 3)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.title('$e^x$')
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.legend()
plt.grid()
plt.show()
```

## We want to approximate $e^x$ around $a = 0$. so the first term of taylor approximation is $f(a) = e^0 = 1$ 

```python
range_of_x = np.linspace(-10, 10, 1000)
y = np.e ** range_of_x

plt.plot(range_of_x, y, label='e^x')

# (f(0) = 1)
taylor_approx_0 = np.ones_like(range_of_x)
plt.plot(range_of_x, taylor_approx_0, label='f(0) = 1', linestyle='--', color='red')

plt.xlim(-1, 1)
plt.ylim(0, 3)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.title('Taylor Series Approximations of $e^x$ around $x=0$')
plt.xlabel('x')
plt.ylabel('$e^x$ and Taylor Approximations')
plt.legend()
plt.grid()
plt.show()
```

## First-order Taylor approximation 
### $f(a) + f'(a)(x - a) = e^0 + e^0 (x - a) = 1 + x$

```python
range_of_x = np.linspace(-10, 10, 1000)
y = np.e ** range_of_x

plt.plot(range_of_x, y, label='e^x')

# (f(0) = 1)
taylor_approx_0 = np.ones_like(range_of_x)
plt.plot(range_of_x, taylor_approx_0, label='f(0) = 1', linestyle='--', color='red')

#(f(0) + f'(0)(x - 0))
a = 0
taylor_approx_1 = 1 + (range_of_x - a)
plt.plot(range_of_x, taylor_approx_1, label='First-Order Approximation', linestyle='--', color='orange')

plt.xlim(-1, 1)
plt.ylim(0, 3)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.title('Taylor Series Approximations of $e^x$ around $x=0$')
plt.xlabel('x')
plt.ylabel('$e^x$ and Taylor Approximations')
plt.legend()
plt.grid()
plt.show()
```

## Second-order Taylor approximation
### $f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 = e^0 + e^0 (x - 0) + \frac{e^0}{2!}(x - 0)^2 =$
### $=1 + x + \frac{x^2}{2}$

```python
range_of_x = np.linspace(-10, 10, 1000)
y = np.e ** range_of_x

plt.plot(range_of_x, y, label='e^x')

# (f(0) = 1)
taylor_approx_0 = np.ones_like(range_of_x)
plt.plot(range_of_x, taylor_approx_0, label='f(0) = 1', linestyle='--', color='red')

#(f(0) + f'(0)(x - 0))
a = 0
taylor_approx_1 = 1 + (range_of_x - a)
plt.plot(range_of_x, taylor_approx_1, label='First-Order Approximation', linestyle='--', color='orange')

#(f(0) + f'(0)(x - 0) + (1/2!)x^2)
taylor_approx_2 = 1 + (range_of_x - a) + (1/2) * (range_of_x - a)**2
plt.plot(range_of_x, taylor_approx_2, label='Second-Order Approximation', linestyle=':', color='magenta')

plt.xlim(-1, 1)
plt.ylim(0, 3)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.title('Taylor Series Approximations of $e^x$ around $x=0$')
plt.xlabel('x')
plt.ylabel('$e^x$ and Taylor Approximations')
plt.legend()
plt.grid()
plt.show()
```

## Third-order Taylor approximation

### $= e^0 + e^0 (x - 0) + \frac{e^0}{2!}(x - 0)^2 + \frac{e^0}{3!}(x - 0)^3 = 1 + x + \frac{x^2}{2} + \frac{x^3}{6}$

## Fourth-order Taylor approximation

## Application in economics

## Cobb douglas production function
$$ Y = K ^\alpha L^{1-\alpha}$$

```python
alpha = 0.3
K = np.linspace(1, 10, 100)
L = np.linspace(1, 10, 100)

K, L = np.meshgrid(K, L)
Y = K**alpha * L**(1 - alpha)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(K, L, Y, cmap='viridis')

ax.set_xlabel('Capital (K)')
ax.set_ylabel('Labor (L)')
ax.set_zlabel('Output (Y)')
ax.set_title('Cobb-Douglas Production Function')

plt.show()
```

## First order taylor approximation of cobb douglas production function (linearized)
$$ Y \approx Y^* + \frac{\partial Y}{\partial K} (K - K^*) + \frac{\partial Y}{\partial L} (L - L^*) $$

```python
# Parameters
alpha = 0.3
K = np.linspace(1, 10, 100)
L = np.linspace(1, 10, 100)

# Steady state values
K_star = 5
L_star = 5
Y_star = K_star**alpha * L_star**(1 - alpha)

#partial derivatives
dY_dK = alpha * (K_star**(alpha - 1)) * (L_star**(1 - alpha))
dY_dL = (1 - alpha) * (K_star**alpha) * (L_star**(-alpha))

#Linearized production function
Y_linearized = Y_star + dY_dK * (K - K_star) + dY_dL * (L - L_star)

K, L = np.meshgrid(K, L)
Y_linearized = Y_star + dY_dK * (K - K_star) + dY_dL * (L - L_star)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(K, L, Y_linearized, cmap='viridis')

ax.set_xlabel('Capital (K)')
ax.set_ylabel('Labor (L)')
ax.set_zlabel('Output (Y)')
ax.set_title('Linearized Cobb-Douglas Production Function')

plt.show()
```

# Taylor Approximation of Ln(x)

### Taylor Series for \( \ln(x) \)

The Taylor series expansion of $\ln(x)$ around \( x = 1 \) is given by:

$\ln(x) = \sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{n} (x - 1)^n$

### Approximations

Zeroth Order (Constant Term): $$ \ln(x) \approx \ln(1) = 0 $$

First Order: $$ \ln(x) \approx \ln(1) + \frac{(-1)^{1+1}}{1} (x - 1) = 0 + (x - 1) = x - 1 $$

Second Order: $$ \ln(x) \approx \ln(1) + \frac{(-1)^{1+1}}{1} (x - 1) + \frac{(-1)^{2+1}}{2} (x - 1)^2 = (x - 1) - \frac{1}{2} (x - 1)^2 $$

Third Order: $$ \ln(x) \approx \ln(1) + \frac{(-1)^{1+1}}{1} (x - 1) + \frac{(-1)^{2+1}}{2} (x - 1)^2 + \frac{(-1)^{3+1}}{3} (x - 1)^3 = (x - 1) - \frac{1}{2} (x - 1)^2 + \frac{1}{3} (x - 1)^3 $$

Fourth Order:

$ \ln(x) \approx \ln(1) + \frac{(-1)^{1+1}}{1} (x - 1) + \frac{(-1)^{2+1}}{2} (x - 1)^2 + \frac{(-1)^{3+1}}{3} (x - 1)^3 + \frac{(-1)^{4+1}}{4} (x - 1)^4 =$ 
$=(x - 1) - \frac{1}{2} (x - 1)^2 + \frac{1}{3} (x - 1)^3 - \frac{1}{4} (x - 1)^4 $


```python
x_range = np.linspace(1, 2, 10000)

ln = np.log
print(ln(np.e)) 

y_values = ln(x_range)

plt.plot(x_range, y_values, label = 'ln(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('ln(x)')
plt.legend(loc='upper left') 

plt.grid()
plt.show()
```

# Zerorth order taylor approximation of ln(x)

```python
x_range = np.linspace(1, 2, 1000)  

ln = np.log

y_values = ln(x_range)

# Taylor series approximations
T0 = np.zeros_like(x_range)  # Zeroth order

plt.plot(x_range, y_values, label='ln(x)', color='black')
plt.plot(x_range, T0, label='Zeroth Order', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Series Approximations of ln(x)')
plt.legend(loc='upper left')  # Adjust legend position

plt.grid()
plt.show()
```

# First order taylor approximation of ln(x)

```python
x_range = np.linspace(1, 2, 1000) 

ln = np.log

y_values = ln(x_range)

T0 = np.zeros_like(x_range)  # Zeroth order
T1 = (x_range - 1)  # First order

plt.plot(x_range, y_values, label='ln(x)', color='black')

# Plot the Taylor series approximations
plt.plot(x_range, T0, label='Zeroth Order', linestyle='--')
plt.plot(x_range, T1, label='First Order', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Series Approximations of ln(x)')
plt.legend(loc='upper left')  

plt.grid() 
plt.show()
```

# Second order taylor approximation of ln(x)

```python
x_range = np.linspace(1, 2, 1000) 

ln = np.log

y_values = ln(x_range)

# Taylor series approximations
T0 = np.zeros_like(x_range)  # Zeroth order
T1 = (x_range - 1)  # First order
T2 = T1 - 0.5 * (x_range - 1)**2  # Second order

plt.plot(x_range, y_values, label='ln(x)', color='black')

# Plot the Taylor series approximations
plt.plot(x_range, T0, label='Zeroth Order', linestyle='--')
plt.plot(x_range, T1, label='First Order', linestyle='--')
plt.plot(x_range, T2, label='Second Order', linestyle='-.')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Series Approximations of ln(x)')
plt.legend(loc='upper left')  

plt.grid() 
plt.show()
```

# Third order taylor approximation of ln(x)

```python
x_range = np.linspace(1, 2, 1000) 

ln = np.log

y_values = ln(x_range)

T0 = np.zeros_like(x_range)  # Zeroth order
T1 = (x_range - 1)  # First order
T2 = T1 - 0.5 * (x_range - 1)**2  # Second order
T3 = T2 + (1/3) * (x_range - 1)**3  # Third order

plt.plot(x_range, y_values, label='ln(x)', color='black')

plt.plot(x_range, T0, label='Zeroth Order', linestyle='--')
plt.plot(x_range, T1, label='First Order', linestyle='--')
plt.plot(x_range, T2, label='Second Order', linestyle='-.')
plt.plot(x_range, T3, label='Third Order', linestyle=':')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Series Approximations of ln(x)')
plt.legend(loc='upper left') 

plt.grid() 
plt.show()
```

# Forth order taylor approximation of ln(x)

```python
x_range = np.linspace(1, 2, 1000)  

ln = np.log

y_values = ln(x_range)

T0 = np.zeros_like(x_range)  # Zeroth order
T1 = (x_range - 1)  # First order
T2 = T1 - 0.5 * (x_range - 1)**2  # Second order
T3 = T2 + (1/3) * (x_range - 1)**3  # Third order
T4 = T3 - (1/4) * (x_range - 1)**4  # Fourth order

plt.plot(x_range, y_values, label='ln(x)', color='black')

plt.plot(x_range, T0, label='Zeroth Order', linestyle='--')
plt.plot(x_range, T1, label='First Order', linestyle='--')
plt.plot(x_range, T2, label='Second Order', linestyle='--')
plt.plot(x_range, T3, label='Third Order', linestyle='--')
plt.plot(x_range, T4, label='Fourth Order', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Series Approximations of ln(x)')
plt.legend(loc='upper left')  # Adjust legend position

plt.grid()
plt.show()
```

# Conclusion: 

## Due to the divergence of higher-order terms, the natural logarithm function ln(x) can be effectively approximated using just the first-order Taylor series expansion.
