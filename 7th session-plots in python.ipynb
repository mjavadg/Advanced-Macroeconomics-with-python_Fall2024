### Understanding the Three-Digit Code
The three-digit code is in the format abc, where:

a is the number of rows in the grid.

b is the number of columns in the grid.

c is the index of the subplot.

For example:

121 means a grid with 1 row and 2 columns, and this subplot is in the 1st position.
122 means a grid with 1 row and 2 columns, and this subplot is in the 2nd position.

```python
import matplotlib.pyplot as plt

fig = plt.figure()

# Create the first subplot in a 2x2 grid
axx1 = fig.add_subplot(221, projection='3d')

# Create the second subplot in a 2x2 grid
axx2 = fig.add_subplot(222, projection='3d')

# Create the third subplot spanning the entire bottom row
axx3 = fig.add_subplot(212)

# Adjust the space between the subplots
plt.subplots_adjust(wspace=0.5, hspace=0.5) 

plt.show()
```

## coub douglas function

$$ U = x^{\alpha} y^{\beta}$$

## Step one: defining the function

```python
def cobb_douglas(x, y):
    alpha = 0.5
    beta = 0.5
    return (x ** alpha) * (y ** beta)
```

```python
cobb_douglas
```

## Step two: meshgriding

```python
x2 = np.linspace(1, 1000, 500, dtype = "int")
y2 = np.linspace(1, 1000, 500, dtype = "int")

X2, Y2 = np.meshgrid(x2, y2)

U = cobb_douglas(X2, Y2)
```

## Step three: plotting

```python
fig = plt.figure(figsize = (12, 6))

ax1 = fig.add_subplot(121, projection = "3d")
ax1.plot_surface(X2, Y2, U, cmap = "autumn")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("utility level")
ax.set_title("utility function")

ax2 = fig.add_subplot(122)
contour = ax2.contour(X2, Y2, U, levels = 5, cmap = "magma")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax.set_title("indifference curves")
plt.colorbar(contour, ax = ax2, label = "utility")

plt.tight_layout()
plt.show()
```
